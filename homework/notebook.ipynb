```python
import pandas as pd
import re

file_path = '../files/input/clusters_report.txt'
with open(file_path, 'r', encoding='utf-8') as file:
    lines = file.readlines()

lines
```

```python
data_lines = lines[4:]
data_lines
```

```python
rows = []
current_row = []
for line in data_lines:
    if re.match(r'^\s*\d+\s+', line):
        if current_row:
            rows.append(current_row)
        current_row = [line.strip()]
    else:
        current_row.append(line.strip())

if current_row:
    rows.append(current_row)

rows
```

```python

processed_rows = []
for row in rows:
        row_data = " ".join(row)  
        parts = re.split(r'\s{2,}', row_data)  
        cluster_number = int(parts[0])  
        keyword_count = int(parts[1]) 
        keyword_percentage = parts[2].replace(',', '.').strip('%')
        keywords = " ".join(parts[3:])
        keywords = re.sub(r'\s+', ' ', keywords)
        keywords = keywords.strip()  
        keywords = keywords.replace(' ,', ',') 
        keywords = keywords.replace(', ', ', ').replace(', ', ', ').strip(', ') 
        keywords = keywords.rstrip('.')
        processed_rows.append({ 
            'cluster': cluster_number,
            'cantidad_de_palabras_clave': keyword_count,
            'porcentaje_de_palabras_clave': float(keyword_percentage),
            'principales_palabras_clave': keywords
        })

processed_rows
```

```python
df = pd.DataFrame(processed_rows)
df.columns = df.columns.str.lower().str.replace(' ', '_')
df
```
